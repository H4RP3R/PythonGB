## Задача №67.  

Создать новый столбец в таблице с пингвинами, который будет отвечать  за показатель длины клюва пингвина.
high - длинный(от 42)
middle - средний(от 35 до 42)
low - маленький(до 35)

```python
df2.loc[df2['bill_length_mm'] >= 42, ['height_group']] = 'high'
df2.loc[(df2['bill_length_mm'] < 42) & (df2['bill_length_mm'] >= 35), ['height_group']] = 'middle'
df2.loc[df2['bill_length_mm'] < 35, ['height_group']] = 'low'
df2.sample(30)
```
